```python
def addVariable():
    print("Variable Declaration")
    print("Array or Variable:")
    arr_or_var = input()
    arr_or_var.lower
    valid_opts = ["array", "variable"]
    while(arr_or_var not in valid_opts):
        print("Not a valid input.")
        print("Array or Variable:")
        arr_or_var = input()
        arr_or_var.lower

    if arr_or_var == "array":
        print("Enter the length of the array:")
        length_arr = input()
        while not(length_arr.isdigit()):
            print("Enter the length of the array:")
            length_arr = input()
    else:
        length_arr = 0

    print("Enter the type of Parameter-(int, float, string):")
    var_type = input()
    valid_var = ["int", "float", "string", "no more"]                
    while(var_type not in valid_var):
        print("Invalid variable type.")
        print("Enter the type of Parameter-(int, float, string):")
        var_type = input()

    print("Enter " + var_type + " variable name:")
    var_name = input()
    var_name.lower()
    test = var_name.split(" ")
    if len(test) != 1:
        var_name = ""
        for i in range(0, len(test)):
            if not(i == len(test) - 1):
                var_name +=  test[i] + "_"
            else:
                var_name += test[i]

    if var_name[0].isdigit():
        print("Not a valid variable name.")
        return -1, -1, -1

    else:
        return var_type, var_name, length_arr
        
def addVariableFunc(type_var, name, length):
    if length != 0:
        statement = type_var +  " " + name + "[" + length + "];\n"

    else:
        statement = type_var + " " + name + ";\n"

    return statement

def funcPrint(name, content):
    addPrintStatement = ("   printf(\"" + content + "\\n\");\n")
    
    return addPrintStatement

def mainVarDeclaration(type_var, name, length, code):
    pointer = 0
    for i in range(0, len(code)):
        if code[i] == "int main(){\n":
            pointer = i
    if length != 0:
        statement = type_var +  " " + name + "[" + length + "];\n"

    else:
        statement = type_var + " " + name + ";\n"

    return pointer, statement

def addFunction(func_name, parameters, return_type, function_input):
    statement = return_type + " " + func_name + "("
    if len(parameters) != 0:
        for i in range(0, len(parameters)):
            if i != len(parameters) - 1:
                statement += parameters[i] + ", "
            else:
                statement += parameters[i] + "){\n"        

    if len(parameters) == 0:
        statement += "){\n" 

    function_input.append(statement)
    function_input.append("\n")

    return function_input
    #Add closing bracket.


def addPrintStatement(content, code):
    code.append("   printf(\"" + content + "\\n\");\n")

# Main function starts.
# Variable Declarations 
all_functions = []
function_parameters = []
reserved_words = ["int", "string", "char", "float", "main", ""]

print("Enter your code in simple language and the code will convert it into C code.")

print("Enter the name of the program you wish to write.")
filename = input()
filename.strip()
filename += ".c"
f = open(filename, "w")

# Initial required library commands and declaration of main function.
code = []
code.append("#include<stdio.h>\n\n")
code.append("int main(){\n")
code.append("\n")

print("Start writing your code. Don't forget to press enter after every command.")

end = -1
while(end == -1):
    print("Main")
    user_input = input()
    
    # Code finish marker.    

    command = user_input.split(" ")
    command = command[0]
    user_input = user_input[(len(command) + 1):]
    content = "" 

    if command == "print":
        user_input = user_input.split(" ")
        print(user_input)
        for i in range(0, len(user_input)):        
            content += user_input[i] + " "
        addPrintStatement(content, code)
        print("Print statement added.")

    elif command == "declareFunction":   
        has_to_return = False
        print("Function")
        function_input = []        
        parameters = []
        print("Function Name:")
        func_name = input()        
        print("Function parameters:")
        var_type = "Nope"
        temp = []

        while(var_type != "no more"):            
            print("Does your function require any parameters(Y or N):")            
            valid_choices = ["y", "n", "Y", "N"]
            param_req = "o"
            while(param_req not in valid_choices):
                param_req = input()

            if param_req == 'y' or param_req == "Y":                                                  
                print("Enter the type of Parameter-(int, float, string):")
                var_type = input()
                valid_var = ["int", "float", "string", "no more"]                
                while(var_type not in valid_var):
                    print("Invalid variable type.")
                    print("Enter the type of Parameter-(int, float, string):")
                    var_type = input()                    

                if var_type == "no more":
                    break            

                print("Enter " + var_type + " variable name:")
                parameter = input()
                parameter.lower()
                test = parameter.split(" ")
                if len(test) != 1:
                    parameter = ""
                    for i in range(0, len(test)):
                         if not(i == len(test) - 1):
                            parameter +=  test[i] + "_"
                         else:
                            parameter += test[i]
                
                if not isinstance(parameter, str):
                    print("Not a valid variable name.")
                temp.append(parameter)
                parameter = var_type + " " + parameter
                parameters.append(parameter)

            else:
                var_type = "no more"

        valid_returns = ["void", "int", "string", "float"]
        print("Enter the return type of the function:")
        return_type = input()
        while(return_type not in valid_returns):
            print("Enter the return type of the function:")
            return_type = input()
        if return_type != valid_returns[0]:
            has_to_return = True

        function_parameters.append(temp)
        function_input = addFunction(func_name, parameters, return_type, function_input)
        all_functions.append(func_name)
            
        print("The function has been declared. Now add the function body. Enter -1 to exit the function.")        
                
        while(True):
            func_input = input()
            func_input = func_input.split(" ")
            content = ""
            for i in range(1, len(func_input)):
                content += func_input[i] + " "

            if func_input[0] == "print":
                function_input.append(funcPrint(func_name, content))
            elif func_input[0] == "declareVariable":
                var_type, var_name, length_arr = addVariable()            
                temp.append(var_name)                
                statement = addVariableFunc(var_type, var_name, length_arr)
                parameters.append(statement)
                function_input.insert(1, "   " + statement)

            elif func_input[0] == "-1":
                correct_return = False
                while not correct_return:
                    if has_to_return:
                        return_vars = []                
                        print("What do you want the function to return?")
                        print("Valid variables to return(if any):")
                        for i in range(0, len(parameters)):
                            if parameters[i].split(" ")[0] == return_type:
                                print(temp[i])
                                return_vars.append(temp[i])                    
                        
                        if return_type == valid_returns[2]:
                            print("Enter the string you wish to return:")                                                        
                        else:
                            print("Enter the number you wish to return:")
                        func_return_value = input()
                        if func_return_value not in return_vars:
                            if func_return_value in temp:
                                print("Int function cannot return another variable type.")
                            else:
                                correct_return = True
                                function_input.append("\n   return " + func_return_value + ";\n")                
                function_input.append("}\n\n")
                print("Function added. Moving to main function.")
                break
            else:
                print("Not an identified command.")
        for i in range(0, len(function_input)):
            code.insert(1, function_input[len(function_input) - 1 - i])

    elif command == "declareVariable":
        var_type, var_name, length_arr = addVariable()
        pointer, statement = mainVarDeclaration(var_type, var_name, length_arr, code)
        code.insert(pointer + 1, "   " + statement)
        print("Variable added.", statement)
       

    elif command  == "-1":
        code.append("\n   return 0;\n")
        code.append("}")
        end = 0            

    else:
        print("Not an identified command. Please check the documentation for the language syntax at https://github.com/OmG2011/Simple_Coding_Language/blob/main/README.md")

    # elif command == "add" or command == "sum" or command == "addition":        
    #     print("Enter the separator between the numbers")
    #     separator = input()
    #     print(type(user_input))
    #     num = user_input.split(separator)
    #     for i in range(1, len(num)):
    #         num[i] = int(num[i])
        
    #     type_var = user_input[0]
    #     variable_declaration(type_var, name, code)
    #     code.append("addFunction(sizeof(num)/sizeof(num[0]), )")
    #     addFunction(num)

print(code)
with open(filename, 'w') as f:
    for line in code:
        f.write(line)        
```
